### What the columns mean and their datatypes

<h5>ID : This is a unique identifier of the accident record.</h5>
<h5>SourceIndicates : source of the accident report (i.e. the API which reported the accident.).</h5>
<h5>TMCA : traffic accident may have a Traffic Message Channel (TMC) code which provides more detailed description of the event.</h5>
<h5>Severity : Showsthe severity of the accident, a number between 1 and 4, where 1 indicates the least impact on traffic (i.e., short delay as a result of the accident) and 4 indicates a significant impact on traffic (i.e., long delay).</h5>
<h5>Start_Time :Shows start time of the accident in local time zone.</h5>
<h5>End_Time : Shows end time of the accident in local time zone.</h5>
<h5>Start_Lat : Shows latitude in GPS coordinate of the start point.</h5>
<h5>Start_Lng : Shows longitude in GPS coordinate of the start point.</h5>
<h5>End_Lat : Shows latitude in GPS coordinate of the end point.</h5>
<h5>End_Lng : Shows longitude in GPS coordinate of the end point.</h5>
<h5>Distance(mi) :The length of the road extent affected by the accident.</h5>
<h5>Description : Shows natural language description of the accident.</h5>
<h5>Number : Shows the street number in address field.</h5>
<h5>Street : Shows the street name in address field.</h5>
<h5>Side : Shows the relative side of the street (Right/Left) in address field.</h5>
<h5>City : Shows the city in address field.</h5>
<h5>County : Shows the county in address field.</h5>
<h5>State : Shows the state in address field.</h5>
<h5>Zipcode : Shows the zipcode in address field.</h5>
<h5>Country : Shows the country in address field.</h5>
<h5>Timezone : Shows timezone based on the location of the accident (eastern, central, etc.).</h5>
<h5>Airport_Code : Denotes an airport-based weather station which is the closest one to location of the accident.</h5>
<h5>Weather_Timestamp : Shows the time-stamp of weather observation record (in local time).</h5>
<h5>Temperature(F) : Shows the temperature (in Fahrenheit).</h5>
<h5>Wind_Chill(F) : Shows the wind chill (in Fahrenheit).</h5>
<h5>Humidity(%) : Shows the humidity (in percentage).</h5>
<h5>Pressure(in) : Shows the air pressure (in inches).</h5>
<h5>Visibility(mi) : Shows visibility (in miles).</h5>
<h5>Wind_Direction : Shows wind direction.</h5>
<h5>Wind_Speed(mph) : Shows wind speed (in miles per hour).</h5>
<h5>Precipitation(in) : Shows precipitation amount in inches, if there is any.</h5>
<h5>Weather_Condition : Shows the weather condition (rain, snow, thunderstorm, fog, etc.).</h5>
<h5>Amenity : A Point-Of-Interest (POI) annotation which indicates presence of amenity in a nearby location.</h5>
<h5>Bump : A POI annotation which indicates presence of speed bump or hump in a nearby location.</h5>
<h5>Crossing : A POI annotation which indicates presence of crossing in a nearby location.</h5>
<h5>Give_Way: A POI annotation which indicates presence of give_way sign in a nearby location.</h5>
<h5>Junction : A POI annotation which indicates presence of junction in a nearby location.</h5>
<h5>No_Exit : A POI annotation which indicates presence of no_exit sign in a nearby location.</h5>
<h5>Railway : A POI annotation which indicates presence of railway in a nearby location.</h5>
<h5>Roundabout : A POI annotation which indicates presence of roundabout in a nearby location.</h5>
<h5>Station : A POI annotation which indicates presence of station (bus, train, etc.) in a nearby location.</h5>
<h5>Stop : A POI annotation which indicates presence of stop sign in a nearby location.</h5>
<h5>Traffic_Calming : A POI annotation which indicates presence of traffic_calming means in a nearby location.</h5>
<h5>Traffic_Signal : A POI annotation which indicates presence of traffic_signal in a nearby location.</h5>
<h5>Turning_Loop : A POI annotation which indicates presence of turning_loop in a nearby location.</h5>
<h5>Sunrise_Sunset : Shows the period of day (i.e. day or night) based on sunrise/sunset.</h5>
<h5>Civil_Twilight : Shows the period of day (i.e. day or night) based on civil twilight.</h5>
<h5>Nautical_Twilight : Shows the period of day (i.e. day or night) based on nautical twilight.</h5>
<h5>Astronomical_Twilight : Shows the period of day (i.e. day or night) based on astronomical twilight.</h5>

<h2> Visualization </h2>

```python
data.columns
```

```python
statedata= data.groupby('State').count().reset_index()
```

```python
statedata[['State','ID']]
statedata=statedata.sort_values(['ID'], ascending=[False])
```

```python
fig = px.bar(statedata, x='State', y='ID',labels={'State':'States','ID':'Number of Accidents'},title='US Statewise Traffic Accidents',color='ID',color_continuous_scale='Bluered_r')
fig.show("png")
```
